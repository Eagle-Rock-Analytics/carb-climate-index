## This notebook generates the Cal-CRAI equations used in various reports

```python
import matplotlib.pyplot as plt
```

```python
equation = r"$\mathit{CalCRAI} = \frac{Gov_i + (A \cdot Soc_{adj} \cdot Gov_i) + (B \cdot BE_{adj} \cdot Gov_i) + (C \cdot NS_{adj} \cdot Gov_i)}{Risk_i}$"

plt.figure(figsize=(6, 2))
plt.text(0.5, 0.5, equation, fontsize=14, ha='center', va='center')
plt.axis('off')

plt.savefig(f'cal_crai_calculation.png', dpi=300)

plt.show()
```
